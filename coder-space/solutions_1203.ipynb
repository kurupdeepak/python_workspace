```python
#two sum 
from typing import  List 
def twoSum(nums: List[int], s: int) -> List[int]:
    map = {}
    for i in range(len(nums)):
        k = nums[i]
        n1 = s - k 
        if n1 in map : 
            return [map[n1],i]
        else:
            map[k] = i 
        i += 1
    return "Not found"

twoSum([2,7,11,15],9)
twoSum([3,2,4],6)
twoSum([3,3],6)
twoSum([3,2],6)

   

```

```python
from typing import Optional
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def create_list(n):
    l = ListNode()
    out = l
    for i in n:
        l.next = ListNode(i)
        l = l.next 
    return out.next

def print_list(l):
    while l : 
        print(l.val , end=" -> ") 
        l = l.next
```
