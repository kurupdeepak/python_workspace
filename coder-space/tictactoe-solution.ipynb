```python
class Game:
    def __init__(self,n):
        self.rows = n
        self.n = n
        self.cols = n 
        # self.board = [[-1 for _ in range(self.cols)] for _ in range(self.rows)]
        self.board = [[-1] * self.cols for _ in range(self.rows)]
        self.total_filled = 0
        self.total_cells = self.rows * self.cols
        self.players = {0:"Player 1",1:"Player 2"}

    def move(self,r,c,p):
        if r >= self.n or c >= self.n :
            return "Invalid move"
        if self.board[r][c] == -1 and self.total_filled < self.total_cells :
            self.board[r][c] = p 
            self.total_filled = self.total_filled + 1
        elif self.total_filled == self.total_cells:
            return "Game is draw"
        
        if self.check_row(r,c,p) or self.check_col(r,c,p) or self.check_diag(r,c,p):
            return f"{p} - {self.players[p]} wins"
        return f"No winner yet"
                

    def check_row(self,r,c,p):
        i = 1
        flag = False
        while i < self.n : 
            if self.board[r + i][c] == p:
                i += 1 
                flag = True
                continue
            else :
                break
        return flag
    
    def check_col(self,r,c,p):
        i = 1
        flag = False
        while i < self.n : 
            if self.board[r][c + i] == p:
                i += 1 
                flag = True
                continue
            else :
                break
        return flag
    
    def check_diag(self,r,c,p):
        i = 1
        flag = False
        while i < self.n : 
            if self.board[r + i][c + i] == p:
                i += 1 
                flag = True
                continue
            else :
                break
        return flag

def test_game():
    game = Game(3)
```
